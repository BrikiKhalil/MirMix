```python
############################## CHOW METHOD FOR RESAMPLING SONGS FROM DIRECTORY #####################################



files = librosa.util.find_files( '/home/khalil/Desktop') ##EXTRACTING FILES FROM A CHOZEN DIRECTORY 
files =list(set(files)) 
####INITIALISING ARRAYS 
Songs = [0] * (len(files)) ### EXTRACTED ARRAY OF SONGS
tempo=[0] * (len(Songs))  ### EXTRACTED BPM
onset_env=[0]*(len(Songs))  
Sample_rate=[0] * (len(files)) #### SAMPLE RATE ARRAY 
Sample=[0]*30  
u=-1         
for i in range(0,len(files)):  #### LOOP IN FILES 
    u=u+1
    Songs[i],Sample_rate[i]=librosa.load( files[i],offset=0 ) #### SONG LOADING 
    tempo_audio, beats_audio = librosa.beat.beat_track(Songs[i], Sample_rate[i]) #### TEMPO EXTRACTION 
    Time_beats_audio=librosa.frames_to_time(beats_audio[:], sr=Sample_rate[i])   #### TIME BEAT EXTRACTION 
    Local_Bpm_audio=[None] * len(Time_beats_audio)                               ####################
    for t in range(1,len(Time_beats_audio)-1):                                   #
        Local_Bpm_audio[t-1]=(60)/(Time_beats_audio[t+1] - Time_beats_audio[t])  #
    Local_Bpm_audio_final=[None]*(len(Local_Bpm_audio)-2)                        #LOCAL BPM DEVOLEPMENT
    for r in range(0,len(Local_Bpm_audio)-1):                                    #
        if Local_Bpm_audio[r]!=None:                                             #
            Local_Bpm_audio_final[r]=Local_Bpm_audio[r]                          ####################
    index = [1,len(Time_beats_audio)-1]
    Time_beats_audio = np.delete(Time_beats_audio,index)                         ## REARRANGING ARRAYS
    y=Local_Bpm_audio_final                                                      ## FOR LATER LINEAR REGRESSION
    x=Time_beats_audio
    Resample=[None] * len(Time_beats_audio)
    k=2                                                                          # K= CARD(PARAMETERS) (INTERCEPT + LOCAL BPM)
    m=0                                                                          # ITERATION VARIABLE FOR DETECTING 
    j=9    # STARTING REGRESSION FROM 10 OBSERVATION                             # REGIM CHANGES
    while True:
        j=j+1                                                                    
        n=len(x)                                                                 # NUMBER OF OBSERVATIONS
        fcrit=scipy.stats.f.ppf(q=1-0.05, dfn=k, dfd=n)                          # F VALUE FOR DETECTING CHANGES
        x_a=x[:j]                                                                ## LEFT REGRESSION AND RIGHT REGRESSION 
        x_b=x[j:]                                                               
        y_a=y[:j]
        y_b=y[j:]                                                                
        X_a = sm.add_constant(x_a) 
        model_a = sm.OLS(y_a, X_a)
        fitted_a = model_a.fit()
        RSS_a=fitted_a.ssr                                                       #SQUARED SUM OF RESIDUAL FOR EACH REGRESSION
        X_b = sm.add_constant(x_b)                                               # NEEDED FOR LATER F STATISTIC
        model_b = sm.OLS(y_b, X_b)
        fitted_b = model_b.fit()
        RSS_b=fitted_b.ssr
        X = sm.add_constant(x) 
        model = sm.OLS(y, X)
        fitted = model.fit()                                                     
        RSS_pooled=fitted.ssr
        F=(((RSS_pooled-RSS_a-RSS_b)/k))/((RSS_a+RSS_b)/(n-2*k))                 ### F_STATISTIC 
    
        if abs(F)> fcrit :                                                       ### CONDITION ON RESAMPLING        
            Resample[m]=x[j]                                                     ## RESAMPLE ARRAY OF TIME WHEN CHANGE DETECTED
            x=x_b
            y=y_b
            m=m+1
        if len(x)<10 or len(y)<10 or len(x_a)<10 or len(y_a)<10 or len(x_b)<10 or len(y_b)<10 or j>len(x)-10   :
            break

    while None in Resample: ### RESAMPLE TIME TREATMENT
        Resample.remove(None)
    Resample.insert(0, 0)  
    #### SAMPLED NEW PART SONGS 
    if len(Resample)==1 : #### IF NO REGIME CHANGE DETECTED 
        Sample[u], sr=librosa.load( files[i],offset=0) 
        u=u+1
    else :       #### IF REGIME CHANGE DETECTED 
        
        for j in range(1,len(Resample)-2):
            Sample[u], sr=librosa.load( files[i],offset=Resample[j],duration=Resample[j+1]-Resample[j])
            u=u+1


Sample_final=[0]*u
for i in range(0,u):
     Sample_final[i]=Sample[i]

hop_length = 256
j=0
for i in range(0,len(Sample_final)):###### SAMPLE DATA TREATMENT
    if (type(Sample_final[i])==np.ndarray):
        j=j+1


Sample_final_final=[0]*j
i=-1
for k in range(0,len(Sample_final)): ### NON ARRAY DATA RAISING EXCEPTION 
    if (type(Sample_final[k]) is np.ndarray):
        i=i+1
        Sample_final_final[i]=Sample_final[k]
       
    

tempo_audio_sample = [0] * len(Sample_final_final) #### TEMPO ARRAY OF ALL SAMPLES 
for i in range(0,len(Sample_final_final)):
    tempo_audio_sample[i], beats_audio_sample = librosa.beat.beat_track(Sample_final_final[i], sr)
    
tempo_audio_sample


            
    
```
